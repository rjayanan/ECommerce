```python
# Data Cleaning
import pandas as pd

file_path = '../DataSet/ecommerce_customer_data_custom_ratios.csv'
ecommerce_data = pd.read_csv(file_path)

ecommerce_data.head(), ecommerce_data.info()

ecommerce_data_cleaned = ecommerce_data.drop(columns=["Age", "Customer Name"])

ecommerce_data_cleaned['Purchase Date'] = pd.to_datetime(ecommerce_data_cleaned['Purchase Date'])

ecommerce_data_cleaned['Returns'] = ecommerce_data_cleaned['Returns'].fillna(0)


```

```python
# Verification of Cleaned Data
ecommerce_data_cleaned.info()

# Check a summary of the data for further insights
ecommerce_data_cleaned.describe(include='all')
```
